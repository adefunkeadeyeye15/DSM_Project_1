## Data Collection

```python
# importing the libraries to make the data available in the working environment
import pandas as pd
loan_data = pd.read_csv("/content/loan_data.csv")
```

```python
# Getting an overview of the dataset by looking through the first 10 rows of the dataset
loan_data.head(10)
```

## Data Analysis

```python
# statistical overview of the dataset
loan_data.describe()
```

```python
# import libraries for data visualization
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
loan_data.dtypes
```

```python
# check for unique values in the 'purpose' column
loan_data["purpose"].unique()
```

```python
# number of each values in the 'purpose' column
loan_data['purpose'].value_counts()
```

```python
# Bar Chart of the 'purpse' column
plt.figure(figsize=(12, 4))   # (creates a new figure speciying the width and height in inches)
sns.countplot(x="purpose",data=loan_data)
plt.title(f'Count of Purpose')
plt.show()
```

```python
# number of each values in the 'not.fully.paid' column
loan_data['not.fully.paid'].value_counts()
```

```python
sns.barplot(data=loan_data, x="not.fully.paid", y = loan_data["not.fully.paid"].count())
plt.title(f'Not fully paid')
plt.show()
```

```python
sns.countplot(x="not.fully.paid",data=loan_data)
plt.title(f'Not fully paid Chart')
plt.show()
```

```python
# number of each values in the 'credit.policy' column
loan_data['credit.policy'].value_counts()
```

```python
# Credit Policy Bar Chart
sns.barplot(data=loan_data, x="credit.policy", y = loan_data["credit.policy"].count())
plt.title(f'Credit Policy Bar Chart')
plt.show()
```

```python
sns.countplot(x="credit.policy",data=loan_data)
plt.title(f'Credit Policy Bar Chart')
plt.show()
```

```python
# Histogram of Interest Rate
sns.histplot(data=loan_data, x="int.rate", kde=True)
plt.title(f'Interest Rate')
```

```python
# Histogram of Revolving Balance
sns.histplot(data=loan_data, x="revol.bal", kde=True)
plt.title("Revolving Balance")
```

```python
# Histogram of Log Annual Income
sns.histplot(data=loan_data, x="log.annual.inc", kde=True)
plt.title("Log Annual Income")
```

```python
# Check for relationship between revolving balance and interest rate using scatter plot
plt.scatter(loan_data["int.rate"], loan_data["revol.bal"])
plt.xlabel("int.rate")
plt.ylabel("revol.bal")
plt.show()
```

```python
# Check for relationship between revolving balance and annual income using scatter plot
plt.scatter(loan_data["log.annual.inc"], loan_data["revol.bal"])
plt.xlabel("log.annual.inc")
plt.ylabel("revol.bal")
plt.show()
```
